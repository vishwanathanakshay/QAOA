```python
# Creates a graph adjacency matrix from a 3-regular graph with random +1,-1 edge weights (nodes must be even).
def G_max_cut_3regular(nodes):
    G = nx.random_regular_graph(3,nodes)
    G = nx.to_numpy_array(G)
    for i in range(nodes):
        for j in range(nodes):
            if i<j:
                if G[i][j] != 0:
                    c = np.random.choice([1,-1], p =[0.5,0.5] )
                    G[i][j] = c
                    G[j][i] = c
    
    return G 
```

```python
# Creates a graph adjacency matrix for SK-model with random +1,-1 edge weights.
def G_SK(nodes):

    G=np.zeros((nodes,nodes))
    
    
    for i in range(nodes):
        for j in range(nodes):
            if i<j:
                c = np.random.choice([1,-1], p = [0.5,0.5]) / np.sqrt(nodes)
                G[i][j] = c 
                G[j][i] = c
    
    
    
    return G
```

```python
# Creates a graphs adjacency matrix for a unweighted connected graph with n nodes and m edges. 
def G_connected(n,m):
    nodes = np.arange(0,n)
    if m >= n*(n-1)/2:
        m = n*(n-1)/2
    if m < n-1:
        m = n-1
    
    G = np.zeros((n,n),dtype = 'float64')
    
    counter = 1
    
    e1 = np.random.choice(nodes,2,replace = False)
    G[e1[0]][e1[1]] = 1
    G[e1[1]][e1[0]] = 1
    counter = 1
    list_node1 = list(e1)
    list_node2 = list(nodes)
    
    while len(list_node1)!=n:
        i = random.choice(list_node1)
        j = random.choice(list_node2)
        if i!=j and G[i][j] !=1:
            G[i][j] = 1 
            G[j][i] = 1
            counter+=1
        if j not in list_node1:
            list_node1.append(j)
            
        
    while counter <m:
        i,j = np.where(G == 0)
        pos = random.randint(0,len(i)-1)
        i = i[pos]
        j = j[pos]
        if i!=j:
            G[i][j] = 1 
            G[j][i] = 1
            counter+=1
    return G 
    
    
        
        
```

```python
def tensor(k):
    t = k[0]
    i = 1
    
    while i < len(k) :
        t = np.kron(t,k[i])
        i+=1
    return t
```

```python
def Graph_to_Hamiltonian(G,n): 
    H = np.zeros((2**n), dtype = 'float64') 
    Z = np.array([1,-1],dtype = 'float64')
    
    for i in range(n):
        j = i+1
        while j<n:
            k = [[1,1]]*n
            k = np.array(k,dtype = 'float64')
                
            if G[i][j] !=0:
                k[i] = Z
                k[j] = Z
                H+= tensor(k)*G[i][j]
                    
            j+=1
        
    return H
```
